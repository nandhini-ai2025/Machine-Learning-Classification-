```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, dtype = int, drop_first = True)
```

```python
dataset
```

```python
dataset["classification_yes"].value_counts()
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
```

```python
independent.shape
```

```python
dependent = dataset[['classification_yes']]
```

```python
dependent 
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'kernel':['linear', 'rbf', 'poly', 'sigmoid'],
             'gamma':['auto', 'scale'],
             'C':[10, 100, 1000, 2000, 3000]}
grid = GridSearchCV(SVC(probability = True), param_grid, refit = True, verbose = 3, cv = 5, n_jobs = -1, scoring = 'f1_weighted')

grid.fit(X_train, y_train)
```

```python
re = grid.cv_results_

grid_pred = grid.predict(X_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_pred)

from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_pred)
```

```python
from sklearn.metrics import f1_score
f1_macro =  f1_score(y_test, grid_pred, average = 'weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

```python
print("The confusion matrix: \n", cm)
```

```python
print("The report: \n", clf_report)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, grid.predict_proba(X_test)[:,1])
```
